## Conventional Air polutants:  'CO_t', 'NH3_t', 'NOx_t', 'PM10_t', 'PM2.5_t', 'SO2_t', 'VOC_t'

```python
import statsmodels.api as sm
y = data['PM10_t']
print(X.shape, y.shape)
```

```python
lm = sm.OLS(y,X).fit()
lm.params
print(lm.summary())
```

```python
# 'Coal TJ', 'NatGase TJ', 'Petrol TJ', 'Bio/Waste TJ','NonFossElec TJ', 'Water Withdrawals Kgal'
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_partregress_grid(lm, fig=fig)
```

```python
fig, ax = plt.subplots(figsize=(12,8))
fig = sm.graphics.influence_plot(lm, ax=ax, criterion="cooks")
```
